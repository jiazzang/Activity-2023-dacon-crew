info()를 통해 확인했듯이 train 데이터셋에는 결측값이 없네요 !

전체 데이터 양상을 보니 int64, float64의 수치형 데이터와 object의 범주형 데이터로 나눌 수 있겠네요. 그럼 데이터의 통계를 확인해볼까요?

<font color=orange>DataFrame.describe()</font>를 사용하면 수치형 데이터에 대해서 평균, 분산, 대표값를 출력해줍니다 !

와, 그러니 int64, float64 dtype으로 저장된 컬럼에 대한 정보가 나오겠네요 👌


```python
# 수치형 컬럼의 통계를 요약해 출력해줍니다.
train.describe()
```

엇, 그러면 나머지 object dtype은 어떻게 하나요?

다 방법이 있죠! 😎

<font color=orange>**DataFrame.describe(include=['object'])**</font>를 사용하면 범주형 데이터에 대해서 범주의 개수(unique), 가장 많이 속해있는 범주명(top), 그 범주에 속하는 데이터 개수(freq)를 출력해준답니다.

```python
# 범주형 컬럼의 통계를 요약해 출력해줍니다.
train.describe(include=['object'])
```

describe 요약 정보를 통해서 아래와 같은 정보를 추론할 수 있겠죠?


- addr_kr(주소): 한 주소에 대해 최대 3220번의 거래 데이터가 존재
- exclusive_use_area(전용면적): 9.26~424.32로 존재(제곱미터로 추정)
- year_of_completion(설립연도): 1961년~2017년에 설립한 아파트
- transaction_year_month(거래연월): 2008년 01월~2017 11월에 거래한 기록
- floor(층수): -4~80 범위에 존재
- transaction_real_price(실거래가): 100~820000에 존재(만원으로 추정)

## 3.3 데이터 필터링

지금까지 데이터가 어떻게 생겼는지 확인했어요.(여기까지 오신 모든 분들께 박수 👏👏👏)  



이제 데이터에 대해 조금 더 깊게 알아보겠습니다.  우리가 가진 데이터의 전체가 아닌 <font color=purple>**데이터의 일부분**</font>만 보고 싶을 때는 어떻게 할까요?



 예를 들어, '부산광역시'의 데이터만 모아서 보고 싶을 때, 혹은 '부산광역시' 아파트의 설립연도, 거래일자만 보고 싶을 때는 어떻게 해야 할까요?

이럴 때 사용하는 것이 **<font color=red>데이터 필터링</font>** 입니다. 데이터 필터링이란 데이터 프레임에서 <font color=blue>특정한 데이터만 골라내는 것</font>을 의미합니다.

데이터 필터링에서 가장 중요한 개념이 <font color=green>**불린 인덱싱(boolean indexing)**</font>인데요.

데이터 프레임에 특정 조건을 기입하면 해당 조건에 맞으면 <font color=blue>True</font>, 틀리면 <font color=red>False</font>를 출력합니다. 

불린 인덱싱은  True인 행들만 출력하여 원하는 조건의 행들만 출력하는 방법을 말합니다.

```python
train['city']== '부산광역시'
```

'city'라는 변수에 '부산광역시'라는 조건을 기입하니, '부산광역시' 조건에 맞는 행 번호에만 True가 출력되는 게 보이시나요?

```python
train[train['city']== '부산광역시']
```

이렇게 해당 조건을 다시 원본 데이터로 감싸면 '부산광역시'에 해당하는 데이터만 출력할 수 있습니다. 참 신기하죠?

불린 인덱싱은 데이터 필터링에서 매우 많이 쓰이는 방법이기 때문에 우리 모두 잊지 않기로 약속해요 :) 💕

판다스에서 많이 사용하는 데이터 필터링 기법에는 <font color=red>**loc**</font>와 <font color=red>**iloc**</font>가 있습니다. <br>


 loc와 iloc는 모두 <font color=green>데이터의 원하는 부분만 볼 때 사용</font>한다는 공통점이 있지만 <font color=blue>loc는 명칭을 기반으로 필터링</font>을 하고, <font color=blue>iloc는 위치를 기반으로 필터링</font>을 한다는 차이점이 있습니다.   

먼저 명칭 기반 필터링을 사용하는 loc에 대해 알아보겠습니다.



<font color=red>**loc**</font>는 <font color=orange>**데이터프레임.loc[필터링할 행의 값(조건), 필터링할 열의 값(범위)]**</font> 형태로 데이터의 원하는 범위를 지정해줍니다. 



loc를 이용해서 위에서 보였던 것처럼 부산광역시 데이터만 출력해보겠습니다.

```python
train.loc[train['city']== '부산광역시', : ]
```

'부산광역시' 데이터만 있는 게 보이시나요? 



여기서 앞뒤에 아무것도 쓰지 않은 <font color=blue>콜론(:)</font>은 모든 행을 가져온다는 뜻입니다.

콜론 앞뒤에 행 번호를 기입하여 보고 싶은 행의 범위를 지정할 수도 있습니다.

(여기서도 불린 인덱싱이 쓰였네요. 속닥속닥 🤫)



이제 한 발 더 나아가서 '부산광역시' 아파트의 설립연도와 거래일자 칼럼만 모아서 보겠습니다. 여러 칼럼을 보고 싶을 때는 리스트 형태로 원하는 칼럼을 써주시면 됩니다. 

```python
train.loc[train['city']== '부산광역시', ['year_of_completion' ,	'transaction_year_month'] ]
```

어떤가요?   loc를 사용하면 내가 원하는 데이터만 자유롭게 볼 수 있겠다는 자신감이 들지 않나요?


다음으로 위치 기반 필터링을 사용하는 iloc에 대해 알아보겠습니다.




<font color=red>**iloc**</font>는 Integer location의 약자로 데이터 프레임 행/열의 <font color=blue>순서</font>를 나타내는 정수를 통해 가져오는 방법입니다.




iloc는 loc와 다르게 <font color=orange>**데이터프레임.iloc[필터링할 행의 번호(숫자 범위), 필터링할 열의 번호(숫자 범위)]**</font> 형태로 데이터의 원하는 범위를 지정해줍니다.

loc를 통해 봤던 것처럼 iloc를 사용하여 아파트의 설립연도와 거래일자를 확인해 보겠습니다.

```python
train.iloc[:,[8,9]]
```

아파트의 설립연도와 거래일자의 칼럼 번호가 각각 8번, 9번이기 때문에 칼럼 위치에 리스트 형태로 숫자 8,9를 기입하면 아파트의 설립일자, 거래일자만 확인할 수 있습니다.




이제 아파트의 설립연도와 거래일자를 10번째 행부터 20번째 행까지 관찰해 보겠습니다.

<font color=purple>(파이썬은 인덱스가 0번부터 시작한다는 점을 주의하세요!)</font>

```python
train.iloc[9:20,[8,9]]
```

그렇다면 이제 loc을 이용해서 데이터를 자유자재로 만져봅시다 !
